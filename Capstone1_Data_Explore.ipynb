```python
r = recs[['TOTSQFT','TOTROOMS','TOTALBTU','TOTALDOL']]
r.head()
```

```python
div = {'REGIONC_1':'Northeast',
       'REGIONC_2':'Midwest',
       'REGIONC_3':'South',
       'REGIONC_4':'West'}

s = recs[['REGIONC']]
```

```python
s.head()
d = pd.get_dummies(s, columns=['REGIONC'])
d.head()
```

```python
d = d.rename(columns=div)
d = d.astype(bool)
d.head()
```

```python
df = r.join(d)
df.head()
#df1 = df[df.Midwest.isnull()]
#df1.empty
```

```python
cols = list(df)
count = 0

for c in cols:
    dfe = df[df[c].isnull()]

    if dfe.empty:
        count += 1
    else:
        print('NaN in ' + c)

    if count == len(cols):
        print('No NaN in DataFrame')
```
